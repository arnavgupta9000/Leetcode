```python
'''
Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)
'''
class ListNode():
    pass


def solve(head):
    if not head or not head.next:
        return head  # Edge cases: empty list or single node

    dummy = ListNode(0)  # Dummy node to handle head swap
    dummy.next = head
    prev = dummy
    curr = head

    while curr and curr.next:
        first = curr
        second = curr.next

        # Swap the pair
        first.next = second.next
        second.next = first
        prev.next = second  # Link previous pair to the swapped nodes

        # Move pointers forward
        prev = first
        curr = first.next  # Move to the next pair
        '''
        the reason this pointer swap works is cause imagine we have 1-2-3-4 then with dummy we have
        0-1-2-3-4-5
        then when we do the swaps we have 
        0-2-1-3-4-5

        but notice that now when we do curr.next we get to 3 and not 2.
        '''

    return dummy.next  # New head is the second node
      
        


'''
just go through with a fast and slow pointer

had to use chatgpt to fix up my code slightly
'''
```
